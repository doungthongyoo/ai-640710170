```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("student.csv")
print("=== ข้อมูลตัวอย่าง ===")
print(df.head())
print("\n=== ขนาดข้อมูล ===")
print(df.shape)
print("\n=== สถิติข้อมูล ===")
print(df.describe(include='all'))
print("\n=== Missing Values ===")
print(df.isna().sum())
```

```python
df['GPA'] = df['GPA'].replace(-1.0, np.nan)
```

```python
print(df.drop_duplicates())
```

```python
if 'GPA' in df.columns:
    df['GPA'] = df['GPA'].replace(-1.0, np.nan)
if 'ชื่อ' in df.columns:
    df['ชื่อ'] = df['ชื่อ'].fillna('ไม่ทราบชื่อ')
if 'อายุ' in df.columns:
    if pd.api.types.is_numeric_dtype(df['อายุ']):
        df['อายุ'] = df['อายุ'].fillna(df['อายุ'].median())
    else:
        df['อายุ'] = pd.to_numeric(df['อายุ'], errors='coerce')
        df['อายุ'] = df['อายุ'].fillna(df['อายุ'].median())
if 'เพศ' in df.columns:
    df['เพศ'] = df['เพศ'].replace({'ช': 'ชาย', 'ญ': 'หญิง', 'Male': 'ชาย', 'Female': 'หญิง'})
    df['เพศ'] = df['เพศ'].fillna('ไม่ระบุ')
if 'น้ำหนัก_kg' in df.columns:
    df['น้ำหนัก_kg'] = pd.to_numeric(df['น้ำหนัก_kg'], errors='coerce')
    df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].fillna(df['น้ำหนัก_kg'].median())
```

```python
print(df[['GPA', 'ชื่อ', 'อายุ', 'เพศ', 'น้ำหนัก_kg']].head(10))
```

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'], errors='coerce')
```

```python
print(df[['วันที่ลงทะเบียน']].head(10))
```

```python
missing_counts = df.isnull().sum()
```

```python
print("\n ค่าหายในแต่ละคอลัมน์:")
for col, count in missing_counts[missing_counts > 0].items():
    percentage = (count / len(df)) * 100
    print(f"   - {col}: {count} ค่า ({percentage:.1f}%)")
```

```python
student_clean = df.dropna()
```

```python
student_clean
```

```python
student_clean['ชื่อ'].fillna('Unknown', inplace=True)

mode_year = student_clean['อายุ'].mode()[0]
student_clean['อายุ'].fillna(mode_year, inplace=True)

student_clean['เพศ'].fillna('Unknown', inplace=True)
student_clean['คณะ'].fillna(student_clean['คณะ'].mode()[0], inplace=True)
student_clean['GPA'].fillna(student_clean['GPA'].mean(), inplace=True)
student_clean['ส่วนสูง_cm'].fillna(student_clean['ส่วนสูง_cm'].mean(), inplace=True)
student_clean['น้ำหนัก_kg'].fillna(student_clean['น้ำหนัก_kg'].mean(), inplace=True)
student_clean['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(student_clean['ชั่วโมงเรียน_ต่อสัปดาห์'].mean(), inplace=True)
```

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์'
}
faculty_mapping
```

```python
df['คณะ'] = df['คณะ'].map(faculty_mapping)
```

```python
student_clean.loc[student_clean['GPA'] < 0, 'GPA'] = np.nan
student_clean.loc[student_clean['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = student_clean['GPA'].mean()
student_clean['GPA'].fillna(mean_gpa, inplace=True)
```
